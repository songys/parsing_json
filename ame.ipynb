```python
contents = []
for fname in flist:
    content = ''.join([x.strip() for x in open(fname, 'rt', encoding='utf-8')])
    jobj = json.loads(content)
    contents = []
    for doc in jobj['document']:
        speakers = doc['metadata']['speaker']
        contact_frequency = doc['metadata']['setting']['contact_frequency']
        utterances = doc['utterance']
        for utterance in utterances:
            row = []
            row += [utterance['form'], utterance['id'], utterance['original_form'], utterance['speaker_id'], utterance['time']]
            for speaker in speakers:
                row += [speaker['age'], speaker['occupation'], speaker['sex'], contact_frequency]
            contents.append(row)
            
    pd.DataFrame(contents).to_excel(fname.replace(".json", ".xlsx"))
```
